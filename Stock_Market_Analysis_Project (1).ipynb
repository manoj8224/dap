```python
# ðŸ“¦ Install Required Libraries (if not installed)
# !pip install yfinance pandas numpy matplotlib seaborn

# ðŸ“š Import Libraries
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ðŸ“Š Set Plot Styles
sns.set(style="darkgrid")
plt.rcParams['figure.figsize'] = (12, 6)

# ðŸ¦ Fetch Stock Data (Example: Apple Inc.)
stock = 'AAPL'
data = yf.download(stock, start='2020-01-01', end='2024-12-31')

# ðŸ§® Add Technical Indicators
data['MA20'] = data['Close'].rolling(window=20).mean()
data['MA50'] = data['Close'].rolling(window=50).mean()
data['Daily Return'] = data['Close'].pct_change()

# ðŸ“‹ Display Head and Summary
print("ðŸ“„ First 5 Rows of Data:")
print(data.head())
print("\nðŸ“Š Summary Statistics:")
print(data.describe())

# ðŸ“ˆ Plot Closing Price
plt.figure()
plt.plot(data['Close'], label='Closing Price')
plt.title(f'{stock} Closing Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()

# ðŸ“‰ Plot Moving Averages
plt.figure()
plt.plot(data['Close'], label='Closing Price')
plt.plot(data['MA20'], label='20-Day MA', linestyle='--')
plt.plot(data['MA50'], label='50-Day MA', linestyle='--')
plt.title(f'{stock} - Moving Averages')
plt.legend()
plt.show()

# ðŸ“Š Plot Daily Return Distribution
plt.figure()
sns.histplot(data['Daily Return'].dropna(), bins=50, kde=True, color='purple')
plt.title(f'{stock} Daily Return Distribution')
plt.xlabel('Daily Return')
plt.ylabel('Frequency')
plt.show()

# ðŸ” Plot Daily Return Over Time
plt.figure()
data['Daily Return'].plot(title=f'{stock} Daily Return Over Time', legend=True)
plt.xlabel('Date')
plt.ylabel('Return')
plt.show()
```
